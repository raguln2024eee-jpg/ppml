```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create time series data (1 year)
dates = pd.date_range(start='2025-01-01', periods=365)

# Generate random temperature data
temps = np.random.normal(loc=30, scale=5, size=365)

# Create DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temps})

# Set Date as index
df.set_index('Date', inplace=True)

# Basic Analysis
print("Summary Statistics:\n")
print(df.describe())

# Monthly average temperature
monthly_avg = df.resample('M').mean()
print("\nMonthly Average:\n", monthly_avg)
# Visualization
# Line plot (daily temperature)
plt.figure()
plt.plot(df.index, df['Temperature'])
plt.title('Daily Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.show()

```
