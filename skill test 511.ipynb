```python
import random

# Generate 20 random student scores out of 100
scores = [random.randint(30, 100) for _ in range(20)]
pass_mark = 50

# 1. Overall class performance
average_score = sum(scores) / len(scores)
highest_score = max(scores)
lowest_score = min(scores)

print(f"Scores: {scores}")
print(f"Average: {average_score:.2f}, Highest: {highest_score}, Lowest: {lowest_score}")

# 2. Pass/fail classification & 3. Grading on a curve
results = []
for score in scores:
    status = "Pass" if score >= pass_mark else "Fail"
    
    # Predefined thresholds for grades
    if score >= 90: grade = 'A'
    elif score >= 80: grade = 'B'
    elif score >= 70: grade = 'C'
    elif score >= 60: grade = 'D'
    else: grade = 'F'
    
    results.append((score, status, grade))

print("\nStudent Results (Score, Status, Grade):")
for res in results:
    print(res)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random step counts for 30 days
np.random.seed(42)
steps = np.random.randint(3000, 15000, size=30)

total_steps = np.sum(steps)
average_steps = np.mean(steps)
max_steps = np.max(steps)
min_steps = np.min(steps)

print(f"Total Steps: {total_steps}, Average: {average_steps:.0f}")
print(f"Max: {max_steps}, Min: {min_steps}")

# Identify high activity days (e.g., > 10000 steps)
high_activity_days = np.where(steps > 10000)[0] + 1
print(f"Days with high activity (>10k steps): {high_activity_days}")

# Optional: Visualize trends over time
plt.plot(range(1, 31), steps, marker='o', linestyle='-')
plt.axhline(y=10000, color='r', linestyle='--', label='High Activity Threshold')
plt.title('Daily Step Count Over a Month')
plt.xlabel('Day')
plt.ylabel('Steps')
plt.legend()
plt.show()
```

```python
import numpy as np

# 30 students, 3 subjects (Math, Science, English)
grades = np.random.randint(40, 100, size=(30, 3))

# Subject-wise statistics (axis=0 computes vertically down the rows)
subject_avg = np.mean(grades, axis=0)
subject_max = np.max(grades, axis=0)
subject_min = np.min(grades, axis=0)

subjects = ['Math', 'Science', 'English']
for i, subj in enumerate(subjects):
    print(f"{subj} -> Avg: {subject_avg[i]:.2f}, Max: {subject_max[i]}, Min: {subject_min[i]}")

# Top performing students based on total scores
total_scores = np.sum(grades, axis=1)
top_student_idx = np.argmax(total_scores)
print(f"\nTop performing student is Student {top_student_idx + 1} with total score {total_scores[top_student_idx]}")
```

```python
def calculate_impedance():
    # Example complex impedance values (Real = Resistance, Imaginary = Reactance)
    # Z = R + jX
    Z1 = complex(50, 20)  # e.g., Resistor + Inductor
    Z2 = complex(30, -10) # e.g., Resistor + Capacitor
    
    # Series Impedance: Z_total = Z1 + Z2
    Z_series = Z1 + Z2
    
    # Parallel Impedance: Z_total = 1 / ( (1/Z1) + (1/Z2) )
    Z_parallel = 1 / ((1 / Z1) + (1 / Z2))
    
    print(f"Impedance 1: {Z1} Ohms")
    print(f"Impedance 2: {Z2} Ohms")
    print(f"Total Series Impedance: {Z_series} Ohms")
    print(f"Total Parallel Impedance: {Z_parallel:.2f} Ohms")

calculate_impedance()
```

```python
import numpy as np

# Two 5x3 matrices (5 days, 3 features: Open, High, Low)
stock1 = np.array([
    [100, 105, 99], [102, 106, 100], [101, 104, 98], [105, 110, 104], [108, 112, 107]
])

stock2 = np.array([
    [95, 98, 93], [96, 100, 95], [94, 97, 92], [97, 101, 96], [99, 105, 98]
])

# Element-wise sum and difference
combined_price = stock1 + stock2
daily_spread = stock1 - stock2

print("Combined Price (Sum):\n", combined_price)
print("\nDaily Spread (Difference):\n", daily_spread)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'], 
        'Price': [1200, 800, 300, 150, 50], 
        'Quantity': [10, 20, 15, 25, 50]}

df = pd.DataFrame(data)

# Operation 1: Describe numerical data
print("1. Data Description:\n", df.describe())

# Operation 2: Add a new column (Total Value)
df['Total Value'] = df['Price'] * df['Quantity']

# Operation 3: Sort by Total Value
df_sorted = df.sort_values(by='Total Value', ascending=False)
print("\n3. Sorted Inventory:\n", df_sorted)

# Operation 4: Filter items with quantity > 15
df_filtered = df[df['Quantity'] > 15]

# Operation 5: Sum total inventory value
total_inventory_value = df['Total Value'].sum()
print(f"\n5. Grand Total Value of Inventory: ${total_inventory_value}")

# Plotting
df.plot(x='Product', y='Quantity', kind='bar', color='skyblue', title='Product Quantities')
plt.ylabel('Quantity')
plt.show()
```
