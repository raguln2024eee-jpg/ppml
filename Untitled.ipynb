```python
import pandas as pd
#Empty DataFrame
df=pd.DataFrame()
print(df)
```

```python
emp=pd.Series(['Parker','John','smith','william'])
id=pd.Series([102,107,109,114])
frame={'Emp':emp,'ID':id}
result=pd.DataFrame(frame)
print("\nseries to Data frame\n")
print(result)
print("\nExtracting one column:\n")
print(result['Emp'])
```

```python
print("\nAdding new column:|n")
result['Age']=pd.Series([35,24,40,38])
print(result)
```

```python
print("\nDeleting one column:\n")
del result['Age']
print(result)
```

```python
print("\n Extracting the third row:\n")
print(result.loc[2])
```

```python
print("\nSlicerows:\n",result[1:3])
```

```python
print("\nDeleting particular row:\n",result.drop(1))
```
