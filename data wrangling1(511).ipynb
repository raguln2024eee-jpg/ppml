```python
import pandas as pd
d1={
    "name":["sai","sub","poo"],
    "age":[20,21,22]
}
d2={
    "qualified":[True,True,False]
}
df1=pd.DataFrame(d1)
print("DF1:\n",df1)
df2=pd.DataFrame(d2)
print("DF2:\n",df2)
```

```python
df1=pd.DataFrame([[11,12],[3,4]])
df2=pd.DataFrame([[5,6],[7,8]])
def myfunc(a,b):
  if(a.sum()>b.sum()):
    return a
  else:
    return b
print(df1)
print(df2)
print(df1.combine(df2,myfunc))
```

```python
import pandas as pd
d1={
    "name":["sai","sub","poo"],
    "age":[20,21,22]
}
d2={
    "name":["sai","sub","poo"],
    "age":[15,16,17]
}
df1=pd.DataFrame(d1)
print("DF1:\n",df1)
df2=pd.DataFrame(d2)
print("DF2:\n",df2)
newdf1=df1.merge(df2,on="name",how='left')
newdf2=df1.merge(df2,on="name",how='right')
print("Left Merge:\n",newdf1)
print("Right Merge:\n",newdf2)
```

```python
# Create DataFrame
df=pd.DataFrame({'team':['A','B','C','D'],
          'points':[90,56,78,88],
          'assist':[12,34,23,44],
          'rebounds':[56,43,55,12]})
#view DataFrame
print(df)
#reshape DataFrame from wide formal to long format
df=pd.melt(df,id_vars='team',value_vars=['points','assist','rebounds'])
#View updated DataFrame
print("\nReshape Dataframe:\n",df)
```
