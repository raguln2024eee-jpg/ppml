```python
import pandas as pd

# Load CSV
df = pd.read_csv('student_scores.csv')

# a. Summary statistics
print("Summary Statistics:\n")
print(df[['Maths','Physics','Chemistry']].describe())

# b. Average score for each student
df['Average'] = df[['Maths','Physics','Chemistry']].mean(axis=1)
print("\nStudent Averages:\n")
print(df[['Name','Average']])

# c. Students scoring below 60 in more than 2 subjects
mask = (df[['Maths','Physics','Chemistry']] < 60).sum(axis=1) > 2
low_scores = df[mask]

if not low_scores.empty:
  print("\nStudents scoring below 60 in more than 2 subjects:\n")
  print(low_scores[['Name']])
else:
  print("\nNo students scoring below 60 in more than 2 subjects.\n")
```
